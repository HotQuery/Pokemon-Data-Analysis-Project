```python
## Using Pandas to analyze a pokemon Data Set
import pandas as pd

df = pd.read_csv('pokemon_data.csv.txt')
```

```python
print(df.head(3))
##.tail instead of head prints bottom 3
```

```python
## Read header names for each column ie Name of pokemon, type, HP..
print(df.columns)

## Read and print a Column
print(df['Name'])
## For the code above, print(df.Name) also works though not really if it has two word names
```

```python
## For reading and printing specified multiple columns:
print(df[['Name','Type 1','HP','Sp. Atk']])
```

```python
## To print one ROW use iloc (int at location)
print(df.iloc[0:3])
##You could also do something like print(df.iloc[1:4]) for MULTIPLE rows
```

```python
## to print a specific location use a COMMA, not a colon 
print(df.iloc[1,2])
```

```python
## Iterating through each row in a dataset:
##note: just using print(index,row) will print everything

for index, row in df.iterrows():
    print(index,row['Name'])
```

```python
## For looking for explicit numerical or textual things inside a column, This prints all Grass type pokemon
df.loc[df['Type 1'] == "Grass"]

## note, NOT using print formats things nicer
```

```python
## For a variety of mathemetical descriptions ie mean, standard dev. :
df.describe()
```

```python
## For sorting values by name use .sort_values() the parameter being which column specified, 
## adding optional param ascending = False starts from Z

df.sort_values('Name', ascending = False)
```

```python
## If you wanted to sort using multiple columns: it sorts Type 1 alphabetically (Thus bug is first) then lowest
## to highest Sp. Atk value

## the ascending = [0,1] specifies that the first collumn sorts A-Z, not Z-A, and that Sp. Atk is sorted from lowest to highest
df.sort_values(['Type 1', 'Sp. Atk'], ascending = [0,1] )
```

```python
## For making CHANGES to a dataset:
## We are going to make a TOTAL stat that encompasses HP, Attack..etc
## an important NOTE is that this value will stay added in the dataframe even if it's commented out, after it has
## been run, unless the data frame value is then dropped

#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
```

```python
df = df.drop(columns=['Total'])
df.head(5)
```

```python
## Another adding column method, the colon with the iloc paramet refers to ALL rows, 4:10 refers to the column values we want to
## add, them being Atk, Def, up to Speed.
df['Total'] = df.iloc[:,4:10].sum(axis=1)

## NOTE if you did axis=0, you would add vertically instead
df.head(5)
```

```python
## For moving collumn location around, we're going to move where Total is:
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

```python
## For Saving our dataframe to a new file
df.to_csv('modified.csv')
```

```python
## Filtering data and creating a new file for it - In this case we look for all Grass AND Poison pokemon with HP above 70
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
```

```python
## The code below resets the indexs of a modified table
new_df = new_df.reset_index()

## And we can save it to a new file here
new_df.to_csv('filtered.csv') 
new_df
```

```python
## Within the loc function, we look for all instances of the stirng 'Mega' within the 'Name' column
## By specifying the ~ (Not) this then negates all those instances and will print the remaining
df.loc[~df['Name'].str.contains('Mega')]
```

```python
## Here we have imported regular expressions - we can use this to sort for all fire and grass pokemon. 
## flags=re.i allows us to ignore capitalziation 
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex = True)]
```

```python
## This code sorts for all pokemons whose name starts with "pi", the ^ before "pi" makes the program only look at the beginning 
## of each string

df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I, regex=True)]
```

```python
## Aggregating Data and using Groupby
## Using groupby, we look at all of the Primary types of each pokemon (Type 1) and are looking at their average stats compared to
## other types

df = pd.read_csv('modified.csv')

df.groupby(['Type 1']).mean()

## We can see that on average, for instance, Dragon type pokemon are almost twice as strong on average than Fairy type pokemon
```

```python
## Here we are sorting by highest defense of each type of pokemon, on average. And putting them from highest to lowest by 
## using ascending = False



df = pd.read_csv('modified.csv')

df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```
